**_Pandas offers approaches to hanadle missing data,of which inlcudes: dropna(), fillna(), interpolate() etc. For this dataset, interpolate() method is chosen to handle missing values._**
**_Interpolation uses mathematical algorithms to estimate missing values based on the values of neighboring data points. This can provide a more accurate representation of the underlying data, especially when the missing values are relatively small. Additionally, interpolation can help to preserve the overall trends and patterns in the data, which can be important for exploratory data analysis.**

### Heatmap for the Dataset

#### *_The plot below shows the correlation heatmap of the entire dataset. This comes in handy when we want to identify pairs of variables that are highly correlated or uncorrelated._* 

```python
data_cols = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']

# Compute correlation matrix
corr_data = cleanData[data_cols].corr()

fig,ax = plt.subplots(figsize=(12, 5)) ## Set figure size

# Plot correlation heatmap
data_heatmap = sns.heatmap(corr_data, annot=True, cmap='coolwarm',ax=ax)

data_heatmap.set_xticklabels(data_heatmap.get_xticklabels(), rotation=45)
```

### **DATA VISUALIZATION FOR EACH COLUMN**

#### Below are data visualizations performed on each of cloumns using histogram and pie chart
####  inference above and below the mean can provide important insights into the distribution and patterns in the data, as well as allow for meaningful comparisons between groups

### Data Visualization for 'Fixed Acidity'

```python
plot1 = sns.histplot(data=cleanData, x='fixed acidity',color='lightgreen')

plot1.set_title("Fixed Acidity Histogram") #title
```

```python
mean1 = cleanData['fixed acidity'].mean()

# Count values above and below mean
above_mean_count = cleanData[cleanData['fixed acidity'] > mean1]['fixed acidity'].count()
below_mean_count = cleanData[cleanData['fixed acidity'] < mean1]['fixed acidity'].count()

# Create pie chart
labels = ['Above Mean', 'Below Mean']
sizes = [above_mean_count, below_mean_count]
colors = ['yellow', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Fixed Acidity Above and Below Mean')
plt.show()
```

### **DATA CORRELATION BETWEEN TWO VARIABLES**

#### Below are data correlation performed on some selected coluums use scatter plot

```python
alcohol = cleanData.loc[:, ['alcohol']]
quality = cleanData.loc[:, ['quality']]
facidity = cleanData.loc[:, ['fixed acidity']]
density = cleanData.loc[:, ['density']]

```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(alcohol, quality, color='r')
ax.set_xlabel('Alcohol')
ax.set_ylabel('Quality')
ax.set_title('Quality VS Alcohol')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(facidity, density, color='b')
ax.set_xlabel('Fixed Acidity')
ax.set_ylabel('Density')
ax.set_title('Fixed Acidity VS Density')
plt.show()
```

#### *the plot below shows that Fixed Acidity and Density are well correlated*
